```python
geo_json_file_loc= 'Data/local-area-boundary.geojson'

# open and extract geo json cmponents
def open_geojson():
    with open(geo_json_file_loc) as json_data:
        d = json.load(json_data)
    return d

def get_gpd_df():
    vancouver_json = open_geojson()
    gdf = gpd.GeoDataFrame.from_features((vancouver_json))
    return gdf

gdf = get_gpd_df()

# Import crimes dataset 
df = pd.read_csv("Data/crimedata_csv_all_years.csv")

# Data Wrangling
df = df[["NEIGHBOURHOOD", "YEAR", "MONTH", "DAY", "HOUR", "MINUTE", "TYPE"]]
```
